```python
import pandas as pd
import joblib
import numpy as np

# ==============================
# LOAD MODELS + FILES
# ==============================

clf = joblib.load('../models/classifier.pkl')
reg = joblib.load('../models/regressor.pkl')
features = joblib.load('../models/features.pkl')

route_stats = pd.read_csv('../models/route_stats.csv')
feature_importance = pd.read_csv('../models/feature_importance.csv')

# ==============================
# LOAD DATA
# ==============================

df = pd.read_csv('../data/final_dataset.csv')

X = df[features]

# ==============================
# PREDICTIONS
# ==============================

# Classification
df['Prediction'] = clf.predict(X)
df['Decision'] = df['Prediction'].map({1: 'BUY', 0: 'WAIT'})

# Confidence
if hasattr(clf, "predict_proba"):
    probs = clf.predict_proba(X)
    df['Confidence'] = probs.max(axis=1)
else:
    df['Confidence'] = None

# ==============================
# PRICE PREDICTION (REAL 🔥)
# ==============================

df['ExpectedPrice'] = reg.predict(X)

# ==============================
# MERGE REAL ROUTE STATS
# ==============================

df = df.merge(
    route_stats,
    on=["Source", "Destination"],
    how="left"
)

# ==============================
# EXTRA METRICS (REAL, NU FAKE)
# ==============================

# diferență între preț estimat și medie
df['Price_vs_Avg_%'] = (
    (df['ExpectedPrice'] - df['price_avg']) / df['price_avg']
) * 100

# etichetare dacă e "good deal"
df['DealQuality'] = np.where(
    df['ExpectedPrice'] < df['price_avg'],
    "GOOD DEAL",
    "EXPENSIVE"
)

# risc bazat pe confidence
df['RiskLevel'] = pd.cut(
    df['Confidence'],
    bins=[0, 0.6, 0.8, 1.0],
    labels=["HIGH", "MEDIUM", "LOW"]
)

# ==============================
# TOP FEATURES (GLOBAL)
# ==============================

top_features = feature_importance.head(5)['Feature'].tolist()
df['TopFeatures'] = ", ".join(top_features)

# ==============================
# SAVE
# ==============================

output_file = '../data/predictions_enriched.csv'
df.to_csv(output_file, index=False)

print("✅ Enriched predictions saved!")

print(df.head())
```
