```python
# =========================================================
# Step 1: Import packages
# =========================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# =========================================================
# Step 2: Load dataset
# =========================================================
df = pd.read_csv('../data/flight_dataset_processed_iata.csv')

print(df.head())
print(df.info())
print(df.isnull().sum())

df = df.dropna()

# =========================================================
# Step 3: Convert datetime
# =========================================================
df['DepartureDatetime'] = pd.to_datetime(df['DepartureDatetime'])
df['ArrivalDatetime'] = pd.to_datetime(df['ArrivalDatetime'])

# =========================================================
# Step 4: Feature engineering (BEFORE encoding)
# =========================================================
df['DepartureHour'] = df['DepartureDatetime'].dt.hour
df['DepartureDay'] = df['DepartureDatetime'].dt.day
df['DepartureMonth'] = df['DepartureDatetime'].dt.month
df['DepartureWeekday'] = df['DepartureDatetime'].dt.weekday

# Route key for stats (IMPORTANT: before encoding)
df['RouteKey'] = df['Source'].astype(str) + "-" + df['Destination'].astype(str)

# =========================================================
# Step 5: Encoding categorical variables
# =========================================================
le_airline = LabelEncoder()
df['Airline'] = le_airline.fit_transform(df['Airline'])

all_airports = pd.concat([df['Source'], df['Destination']]).unique()
le_airports = LabelEncoder()
le_airports.fit(all_airports)

df['Source'] = le_airports.transform(df['Source'])
df['Destination'] = le_airports.transform(df['Destination'])

# =========================================================
# Step 6: Scaling
# =========================================================
scaler = MinMaxScaler()
df[['DurationMinutes', 'DepartureHour']] = scaler.fit_transform(
    df[['DurationMinutes', 'DepartureHour']]
)

# =========================================================
# Step 7: ROUTE STATISTICS (FIXED OUTPUT FORMAT)
# =========================================================
stats = df.groupby(['Source', 'Destination']).agg(
    price_min=('Price', 'min'),
    price_max=('Price', 'max'),
    price_avg=('Price', 'mean'),
    price_median=('Price', 'median'),
    samples=('Price', 'count')
).reset_index()

stats = stats.sort_values(by='samples', ascending=False)

print("\n📊 ROUTE STATS:")
print(stats.head(10))

# =========================================================
# Step 8: Merge stats back (optional if needed for ML)
# =========================================================
df = df.merge(stats, on=['Source', 'Destination'], how='left')

# =========================================================
# Step 9: Label creation (BUY / WAIT)
# =========================================================
df['Label'] = np.where(
    df['Price'] <= df['price_min'] + (df['price_max'] - df['price_min']) * 0.25,
    1,  # BUY
    np.where(
        df['Price'] >= df['price_min'] + (df['price_max'] - df['price_min']) * 0.75,
        0,  # WAIT
        -1  # NEUTRAL
    )
)

df = df[df['Label'] != -1]

print("\nDistribuție Label:")
print(df['Label'].value_counts())

# =========================================================
# Step 10: Drop unused columns
# =========================================================
df = df.drop(
    columns=[
        'DepartureDatetime',
        'ArrivalDatetime'
    ],
    errors='ignore'
)

# =========================================================
# Step 11: Visualizations
# =========================================================
plt.figure(figsize=(8,5))
sns.histplot(df['Price'], kde=True)
plt.title("Distribuția prețurilor")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='TotalStops', y='Price', data=df)
plt.title("Preț vs Escale")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x='Airline', y='Price', data=df)
plt.title("Preț vs Companie")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='Label', data=df)
plt.title("BUY vs WAIT")
plt.show()

# =========================================================
# Step 12: Save final dataset
# =========================================================
df.to_csv('../data/final_dataset.csv', index=False)

stats.to_csv('../data/route_stats.csv', index=False)

print("\n✅ Dataset + stats salvate corect!")
```
