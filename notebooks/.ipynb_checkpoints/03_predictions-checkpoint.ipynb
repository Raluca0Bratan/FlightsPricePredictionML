```python
import pandas as pd
import joblib

# ======================
# LOAD
# ======================
reg = joblib.load("../models/regressor.pkl")
clf = joblib.load("../models/classifier.pkl")
features = joblib.load("../models/features.pkl")

df = pd.read_csv("../data/final_dataset.csv")

X = df[features]

# ======================
# PREDICT
# ======================
df["ExpectedPrice"] = reg.predict(X)
df["ML_Decision"] = clf.predict(X)

# ======================
# SMART DECISION ENGINE
# ======================
def final_decision(row):
    if row["ML_Decision"] == 1 and row["ExpectedPrice"] < row["price_avg"]:
        return "🔥 STRONG BUY"
    elif row["ML_Decision"] == 1:
        return "BUY"
    elif row["ExpectedPrice"] > row["price_avg"] * 1.1:
        return "❌ OVERPRICED"
    else:
        return "WAIT"

df["FinalDecision"] = df.apply(final_decision, axis=1)

# ======================
# SAVE
# ======================
df.to_csv("../data/predictions.csv", index=False)

print(df["FinalDecision"].value_counts())
print("✅ Predictions DONE")
```
