```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# ======================
# LOAD
# ======================
df = pd.read_csv('../data/full_combinations_flights.csv')
df.columns = df.columns.str.strip().str.lower()
df = df.dropna()

print("Columns:", df.columns.tolist())

# ======================
# FIX TOTAL STOPS (ROBUST)
# ======================
def get_column(df, names):
    for n in names:
        if n in df.columns:
            return df[n]
    raise Exception(f"Missing column: {names}")

df["Total_Stops"] = get_column(df, ["total_stops", "totalstops", "total stops"])

# ======================
# DATETIME
# ======================
df['departuredatetime'] = pd.to_datetime(df['departuredatetime'])
df['arrivaldatetime'] = pd.to_datetime(df['arrivaldatetime'])

df['DepartureHour'] = df['departuredatetime'].dt.hour
df['DepartureDay'] = df['departuredatetime'].dt.day
df['DepartureMonth'] = df['departuredatetime'].dt.month
df['DepartureWeekday'] = df['departuredatetime'].dt.weekday

df["IsWeekend"] = df["DepartureWeekday"].isin([5,6]).astype(int)

# 🔥 IMPORTANT FEATURE
today = df['departuredatetime'].min()
df["DaysUntilDeparture"] = (df['departuredatetime'] - today).dt.days

# ======================
# ENCODING
# ======================
le_airline = LabelEncoder()
df['airline'] = le_airline.fit_transform(df['airline'])

all_airports = pd.concat([df['source'], df['destination']]).unique()
le_airports = LabelEncoder()
le_airports.fit(all_airports)

df['source'] = le_airports.transform(df['source'])
df['destination'] = le_airports.transform(df['destination'])

# ======================
# SCALING
# ======================
scaler = MinMaxScaler()
df[['durationminutes', 'DepartureHour']] = scaler.fit_transform(
    df[['durationminutes', 'DepartureHour']]
)

# ======================
# ROUTE STATS
# ======================
stats = df.groupby(['source', 'destination']).agg(
    price_min=('price', 'min'),
    price_max=('price', 'max'),
    price_avg=('price', 'mean')
).reset_index()

df = df.merge(stats, on=['source', 'destination'], how='left')

# ======================
# LABEL (BUY / WAIT)
# ======================
df['Label'] = np.where(
    df['price'] <= df['price_avg'] * 0.9, 1, 0
)

# ======================
# SAVE
# ======================
df.to_csv('../data/final_dataset.csv', index=False)
stats.to_csv('../data/route_stats.csv', index=False)

print("✅ Preprocessing DONE")
```
