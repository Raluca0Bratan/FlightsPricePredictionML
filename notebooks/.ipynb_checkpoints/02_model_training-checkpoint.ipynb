```python
import pandas as pd
import joblib
import os

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score

import matplotlib.pyplot as plt
import shap

# ======================
# LOAD
# ======================
df = pd.read_csv("../data/final_dataset.csv")

# ======================
# FEATURES
# ======================
features = [
    "airline",
    "source",
    "destination",
    "Total_Stops",
    "durationminutes",
    "DepartureHour",
    "DepartureDay",
    "DepartureMonth",
    "DepartureWeekday",
    "IsWeekend",
    "DaysUntilDeparture",
    "price_avg",
    "price_min",
    "price_max"
]

X = df[features]

# ======================
# REGRESSION
# ======================
y_reg = df["price"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y_reg, test_size=0.2, random_state=42
)

reg = RandomForestRegressor(n_estimators=200, max_depth=14)
reg.fit(X_train, y_train)

pred = reg.predict(X_test)

print("REGRESSION:")
print("MAE:", mean_absolute_error(y_test, pred))
print("R2:", r2_score(y_test, pred))

# ======================
# CLASSIFIER (BUY/WAIT)
# ======================
y_clf = df["Label"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y_clf, test_size=0.2, random_state=42
)

clf = RandomForestClassifier(n_estimators=200)
clf.fit(X_train, y_train)

pred = clf.predict(X_test)

print("\nCLASSIFIER:")
print("Accuracy:", accuracy_score(y_test, pred))

# ======================
# FEATURE IMPORTANCE
# ======================
importances = reg.feature_importances_

plt.figure()
plt.barh(features, importances)
plt.title("Feature Importance")
plt.show()

# ======================
# SHAP (EXPLAINABILITY)
# ======================
explainer = shap.TreeExplainer(reg)
shap_values = explainer.shap_values(X.sample(100))

shap.summary_plot(shap_values, X.sample(100))

# ======================
# SAVE
# ======================
os.makedirs("../models", exist_ok=True)
joblib.dump(le_airline, "../models/le_airline.pkl")
joblib.dump(le_airports, "../models/le_airports.pkl")
joblib.dump(scaler, "../models/scaler.pkl")

joblib.dump(reg, "../models/regressor.pkl")
joblib.dump(clf, "../models/classifier.pkl")
joblib.dump(features, "../models/features.pkl")

print("✅ Models saved")
```
